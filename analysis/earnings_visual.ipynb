```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
# plots will display inline
%matplotlib inline 

recent_grads = pd.read_csv('../data/recent_grads.csv')

recent_grads.info()
recent_grads.describe()

# drop rows with null values
recent_grads = recent_grads.dropna(axis=0)

#recent_grads.info()


```

### Panda: Scatter Plot

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median', figsize=(5,5))
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter',title='Sample_size vs. Unemployment_rate', figsize=(5,5))
recent_grads.plot(x='Full_time', y='Median', kind='scatter',title='Full_time vs. Median', figsize=(5,5))
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter',title='ShareWomen vs. Unemployment_rate', figsize=(5,5))
recent_grads.plot(x='Men', y='Median', kind='scatter',title='Men vs. Median', figsize=(5,5))
recent_grads.plot(x='Women', y='Median', kind='scatter',title='Women vs. Median', figsize=(5,5))
```

### Panda: Bar Chart - distribution of values in a column

```python
# Series.plot() to plot a specific column
# DataFrame.plot() to generate plots that use values from multiple columns. 

cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,24))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=5,title=cols[r])


## series.hist can be used to change "bin" 
## recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(8,24))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40, title=cols[r])
```

### Panda: scatter matrix plot combines both scatter plots and histograms into one grid of plots

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))



```

### Panda:  bar plot

```python
recent_grads_sort_share_women = recent_grads.sort_values(by=["ShareWomen"],ascending=False)
recent_grade_sort_unemployment =  recent_grads.sort_values(by=["Unemployment_rate"],ascending=False)



recent_grads_sort_share_women[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads_sort_share_women[163:].plot.bar(x='Major', y='ShareWomen', legend=False)

recent_grade_sort_unemployment [:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grade_sort_unemployment [163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)

# matplotlib: recent_grads[:5].plot.bar(x='Major', y='Women')
```
